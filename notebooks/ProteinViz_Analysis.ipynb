# 🧬 ProteinViz Analysis

A structural bioinformatics project developed using Python and Biopython.

This notebook demonstrates the analyses performed on the Acetylcholinesterase (AChE) protein structure (PDB ID: 4M0E).

## Import Libraries

```python
from Bio.PDB import PDBParser
import pandas as pd
import matplotlib.pyplot as plt
```

## Load the Protein Structure

```python
parser = PDBParser(QUIET=True)
structure = parser.get_structure("AChE", "../data/proteins/4M0E.pdb")

print("Protein ID:", structure.id)
print("Chains:", len(list(structure.get_chains())))
print("Residues:", len([r for r in structure.get_residues() if r.id[0] == " "]))
```
